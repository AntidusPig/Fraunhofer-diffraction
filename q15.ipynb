```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as smp
```

```python
z1,z2,z3,k2,l = smp.symbols("Z_1,Z_2,Z_3,k_2,l")
```

```python
A = smp.Matrix([[-1,1,1,0],[1,z1/z2,-z1/z2,0],[0,smp.exp(-1j*k2*l),smp.exp(1j*k2*l),-1],[0,smp.exp(-1j*k2*l),-smp.exp(1j*k2*l),-z2/z3]])
b = smp.Matrix([1,1,0,0])
b.transpose()
```

```python
A
```

```python
b
```

```python
x = smp.MatMul(A.inv(),b)
```

```python
x.transpose()
```

```python
l_subs = np.linspace(0,1,100)
exp = x.evalf(subs={k2:2*smp.pi, z1:1, z2:2, z3:4})
```

```python
exp = smp.simplify(str(exp))
```

```python
result = np.array([exp.evalf(subs={l:l_subs[i]}) for i in range(len(l_subs))],dtype=complex)
```

```python
r,tau = result[:,0],result[:,3]
```

```python
r.dtype
```

```python
fig,(ax1,ax2) = plt.subplots(1,2)
fig.tight_layout()
ax1.plot(l_subs, np.abs(r), label=r'$|r|$')
ax2.plot(l_subs, np.angle(r), label=r'$\arg r$')
ax1.plot(l_subs, np.abs(tau), label=r'$|\tau|$')
ax2.plot(l_subs, np.angle(tau), label=r'$\arg \tau$')
ax1.set_xlabel(r'$l$')
ax1.set_ylabel(r'$amplitude$')
ax2.set_xlabel(r'$l$')
ax2.set_ylabel(r'$angle$')
ax1.legend()
ax2.legend()
```

```python
exp = x.evalf(subs={k2:2*smp.pi, z1:1, z2:3, z3:4})
exp = smp.simplify(str(exp))
result = np.array([exp.evalf(subs={l:l_subs[i]}) for i in range(len(l_subs))],dtype=complex)
r,tau = result[:,0],result[:,3]
fig,(ax1,ax2) = plt.subplots(1,2)
fig.tight_layout()
ax1.plot(l_subs, np.abs(r), label=r'$|r|$')
ax2.plot(l_subs, np.angle(r), label=r'$\arg r$')
ax1.plot(l_subs, np.abs(tau), label=r'$|\tau|$')
ax2.plot(l_subs, np.angle(tau), label=r'$\arg \tau$')
ax1.set_xlabel(r'$l$')
ax1.set_ylabel(r'$amplitude$')
ax2.set_xlabel(r'$l$')
ax2.set_ylabel(r'$angle$')
ax1.legend()
ax2.legend()
```
